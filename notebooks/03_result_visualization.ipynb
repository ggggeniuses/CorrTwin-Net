# Tutorial: CorrTwin-Net Result Visualization

This notebook loads generated result figures and metric files.

```python
from __future__ import annotations

import json
from pathlib import Path

from IPython.display import Image, display

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
FIGURES = ROOT / 'results/figures'
METRICS = ROOT / 'results/metrics'
FIGURES, METRICS
```

## Prediction Curves

```python
for name in ['mlp_acf_prediction_example.png', 'mlp_fcf_prediction_example.png']:
    path = FIGURES / name
    if path.exists():
        display(Image(filename=str(path)))
    else:
        print(f'Missing {path}; run experiments/run_evaluate_curves.py first.')
```

## Metrics

```python
for path in sorted(METRICS.glob('*metrics.json')):
    print(path.name)
    print(json.dumps(json.loads(path.read_text(encoding='utf-8')), indent=2))
```
