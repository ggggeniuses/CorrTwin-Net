## Inspect One Label

`y` is the concatenation of normalized ACF and FCF curves.

```python
import matplotlib.pyplot as plt

curve_points = 64
plt.figure(figsize=(7, 4))
plt.plot(y[0, :curve_points], label='ACF')
plt.plot(y[0, curve_points:], label='FCF')
plt.ylim(-0.05, 1.05)
plt.grid(alpha=0.25)
plt.legend();
```
