# Tutorial: CorrTwin-Net Surrogate Training

This notebook shows the command-line training workflow used by the project.

```python
from __future__ import annotations

import subprocess
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
ROOT
```

## Small Reproducible Run

The commands below are intentionally small. Increase `--samples` and `--epochs` for final experiments.

```python
commands = [
    [sys.executable, 'experiments/run_generate_dataset.py', '--samples', '120', '--curve-points', '64'],
    [sys.executable, 'experiments/run_train_mlp.py', '--epochs', '5', '--batch-size', '32', '--device', 'cpu'],
]

for cmd in commands:
    print('+', ' '.join(cmd))
    subprocess.run(cmd, cwd=ROOT, check=True)
```
